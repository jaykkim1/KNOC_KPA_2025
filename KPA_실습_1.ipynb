```python
# 수포화도 계산 함수
def archie_sw(Rt, Rw, phi, a, m, n):
    """
    Archie 방정식으로 수포화도 계산

    Parameters:
    Rt : float - true resistivity (ohm-m)
    Rw : float - formation water resistivity (ohm-m)
    phi : float - porosity (0~1 사이)
    a : float - tortuosity factor
    m : float - cementation exponent
    n : float - saturation exponent

    Returns:
    Sw : float - 수포화도 (0~1 사이)
    """
    if phi <= 0 or Rt <= 0 or Rw <= 0:
        raise ValueError("공극률, Rt, Rw 값은 모두 양수여야 합니다.")

    Sw = ((a / (phi ** m)) * (Rw / Rt)) ** (1 / n)
    return Sw


Rt = 10.25
Rw= 0.05
phi= 0.2
a = 1
m = 2
n = 2

archie_sw(Rt, Rw, phi, a, m, n)
```

```python
import numpy as np
import pandas as pd

def RT_calc(a, RW, PHI, SW, m, n):
  RT = (a*RW)/((PHI**m)*(SW**n))
  RT = round(RT,3)
  return RT

a = 1
RW = 0.018
PHI = 0.14
SW = 0.4
m = 1.94
n = 2.03


PHI_range = np.arange(0.11,0.16,0.005)
SW_range = np.arange(0.1,1.1,0.1)
result = pd.DataFrame(index=PHI_range, columns=SW_range)

for i in PHI_range:
  for j in SW_range:
    # print(RT_calc(a, RW, i, j, m, n))

    result.loc[i,j] = round(RT_calc(a, RW, i, j, m, n),3)

print("RT sensitivity w/ SW(column) & PHIT(row)")
result
```
